```python
# para elementos sueltos
def FreeElements(BiMatrix):
    elements=set(BiMatrix[BiMatrix.iloc[0:,len(BiMatrix.columns)-1]==1].iloc[0:,0])
    for i in range(1,len(BiMatrix.columns)-1):
        elements=elements.difference(set(BiMatrix[BiMatrix.iloc[0:,i]==1].iloc[0:,0]))
        set(BiMatrix[BiMatrix.iloc[0:,i]==1].iloc[0:,0]).union(set(BiMatrix[BiMatrix.iloc[0:,i+1]==1].iloc[0:,0]))
    return list(elements)
```
